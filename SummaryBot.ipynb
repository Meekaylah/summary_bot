```python
import os
from dotenv import load_dotenv
from slack_sdk import WebClient
from slack_sdk.errors import SlackApiError

load_dotenv()

# Initialize Slack client
slack_token = os.getenv('SLACK_TOKEN')
client = WebClient(token=slack_token)

# Define the channel ID
channel_id = os.getenv('CHANNEL_ID')

try:
    # Fetch last 100 messages
    response = client.conversations_history(
        channel=channel_id,
        limit=100
    )
    messages = response["messages"]
    
    for message in messages:
        print(message["text"])
        
except SlackApiError as e:
    print(f"Error fetching messages: {e.response['error']}")
```

```python
text_to_summarize = " ".join([message["text"] for message in messages])
```

```python
from transformers import pipeline

# Load summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Summarize the text
try:
    summary = summarizer(text_to_summarize, max_length=300, min_length=30, do_sample=False)
    print(summary[0]['summary_text'])
except Exception as e:
    print(f"Error summarizing text with Hugging Face model: {e}")
```

```python
try:
    response = client.chat_postMessage(
        channel=channel_id,
        text=f"*Lemme tell you the gist:*\n\n{summary[0]['summary_text']}"
    )
except SlackApiError as e:
    print(f"Error posting message: {e.response['error']}")
```
